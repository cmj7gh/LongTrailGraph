```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()

G.add_node("A",pos=(0,0))
G.add_node("B",pos=(4.5,0))
G.add_node("C",pos=(31.5,0))
G.add_node("D",pos=(36,0))
G.add_node("E",pos=(4.5,18))
G.add_node("F",pos=(18,18))
G.add_node("G",pos=(31.5,18))
G.add_node("H",pos=(4.5,60))
G.add_node("I",pos=(18,60))
G.add_node("J",pos=(31.5,60))
G.add_node("K",pos=(0,78))
G.add_node("L",pos=(4.5,78))
G.add_node("M",pos=(31.5,78))
G.add_node("N",pos=(36,78))

pos=nx.get_node_attributes(G,'pos')


G.add_edge("A","B")
G.add_edge("B","C")
G.add_edge("C","D")
G.add_edge("A","K")
G.add_edge("B","E")
G.add_edge("C","G")
G.add_edge("D","N")
G.add_edge("E","H")
G.add_edge("F","I")
G.add_edge("G","J")
G.add_edge("E","F")
G.add_edge("F","G")
G.add_edge("H","I")
G.add_edge("I","J")
G.add_edge("H","L")
G.add_edge("J","M")
G.add_edge("K","L")
G.add_edge("L","M")
G.add_edge("M","N")

nx.draw(G,pos,with_labels=True)
```

```python
from gurobipy import *

# Create a new model
m = Model()

# Create variables
AB = m.addVar(vtype=GRB.BINARY, name="AB")
BC = m.addVar(vtype=GRB.BINARY, name="BC")
CD = m.addVar(vtype=GRB.BINARY, name="CD")
AK = m.addVar(vtype=GRB.BINARY, name="AK")
BE = m.addVar(vtype=GRB.BINARY, name="BE")
CG = m.addVar(vtype=GRB.BINARY, name="CG")
DN = m.addVar(vtype=GRB.BINARY, name="DN")
EH = m.addVar(vtype=GRB.BINARY, name="EH")
FI = m.addVar(vtype=GRB.BINARY, name="FI")
GJ = m.addVar(vtype=GRB.BINARY, name="GJ")
EF = m.addVar(vtype=GRB.BINARY, name="EF")
FG = m.addVar(vtype=GRB.BINARY, name="FG")
HI = m.addVar(vtype=GRB.BINARY, name="HI")
IJ = m.addVar(vtype=GRB.BINARY, name="IJ")
HL = m.addVar(vtype=GRB.BINARY, name="HL")
JM = m.addVar(vtype=GRB.BINARY, name="JM")
KL = m.addVar(vtype=GRB.BINARY, name="KL")
LM = m.addVar(vtype=GRB.BINARY, name="LM")
MN = m.addVar(vtype=GRB.BINARY, name="MN")

#objective function
m.setObjective(AB * 4.5 + BC * 27 + CD * 4.5 + AK * 78 + BE * 18 + CG *18 + DN * 78 
               + EH * 42 + FI * 42 + GJ * 42 + EF * 13.5 + FG * 13.5 + HI * 13.5 + IJ * 13.5 
               + HL * 18 + JM * 18 + KL * 4.5 + LM * 27 + MN * 4.5, GRB.MINIMIZE)

#constraints for nodes that are currently odd. 
#We need them to be even, so the total number of edges added to each of these nodes must be an odd number
m.addConstr((AB + BC + BE) == 1, "c0")
m.addConstr((BE + EF + EH) == 1, "c2")
m.addConstr((EF + FI + FG) == 1, "c3")
m.addConstr((FG + GJ + CG) == 1, "c4")
m.addConstr((EH + HI + HL) == 1, "c5") 
m.addConstr((HI + FI + IJ) == 1, "c6")
m.addConstr((IJ + GJ + JM) == 1, "c7")
m.addConstr((LM + JM + MN) == 1, "c9")

#It doesn't need to be a closed loop, so two nodes can stay odd
#m.addConstr((HL + LM + KL) == 1, "c8")
#m.addConstr((BC + CG + CD) == 1, "c1")

#constraints for nodes that are currently even. 
#We need them to stay even, so the total number of edges added to each of these nodes must be an even number
m.addConstr((AB + AK) == 0, "C10")
m.addConstr((AK + KL) == 0, "C11")
m.addConstr((CD + DN) == 0, "C12")
m.addConstr((MN + DN) == 0, "C13")


#do the thing
m.optimize()

#print the results
for v in m.getVars():
    print('%s %g' % (v.varName, v.x))

print('Obj: %g' % m.objVal)
```
